# 1. Laad Library's

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

# 2. Laad Image

```python
zebra = imread('zebrapad01.jpg')
```

# 3. Plot image

```python
plt.figure(figsize=(10,10))
plt.imshow(zebra)
```

# 4. Prepareer data

Een K-Means model kan alleen maar getallen tussen de 0 en 1 krijgen. Dus elk getal /255 maakt dit mogelijk.

```python
zebra = zebra/255.0
```

```python
zebra.shape
```

Het model verwacht ook een andere shape. Dus dat moet ook gebeuren met reshape()

```python
x = zebra.reshape(-1,3)
```

```python
x.shape
```

# 5. Image Segmentatie

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)
km.fit(x)
```

Zoals vermeld heb ik een n_clusters van 3 ingesteld. Dus ik verwacht een afbeelding wat 3 kleuren heeft.

```python
img_seg = km.cluster_centers_[km.labels_].reshape(zebra.shape)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img_seg)
```
