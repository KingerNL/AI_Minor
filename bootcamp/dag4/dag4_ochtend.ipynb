## Zet een kleurenplaatje om in een grijswaarde plaatje.

Als je foto al zwart-wit is dan kun je deze stap overslaan.

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

## Initieer detector

Download de .xml file met de classifier: haarcascade_frontalface_default.xml

Initieer de Cascade Classifier.

```python
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')

print("Is de classifier leeg? --> " + str(face_cascade.empty()) + "  (moet 'False', dus 'niet leeg' zijn)")
```

## Detecteer gezichten

Bepaal met de classifier de gezichten in het plaatje.

Hoeveel heb je er gevonden?

```python
faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(10, 10),
        flags=cv.CASCADE_SCALE_IMAGE)

print("Aantal gezichten: " + str(faces.shape[0]))
```

## Toon rechthoeken rond gezichten

Toon rechthoeken om de gevonden gezichten.

```python
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(25,100,100),2)
    
plt.figure(figsize=(15,10))
plt.imshow(img)
plt.axis('off');
```

Je kan het ook als zwart-wit plaatje laten zien.

```python
for (x,y,w,h) in faces:
    cv.rectangle(gray,(x,y),(x+w,y+h),(255,0,0),1)
    
plt.figure(figsize=(15,10))
plt.imshow(gray, cmap='gray') # colormap aanpassen!
plt.axis('off');
```
