## 1. Wat is diabetes?


---

### Wat is diabetes, en zijn er verschillende typen? 


> Diabetes is een ziekte waarbij het lichaam de bloedsuiker niet meer in evenwicht kan houden. Dat komt doordat het lichaam weinig insuline heeft. Ook regeert het lichaam vaak niet meer goed op insuline of wordt het helemaal niet meer aangemaakt.



### Hoe wordt de diagnose gesteld? 
> Diabetes kan worden vastgesteld doormiddel van bloedonderzoek. Daar wordt dan gekeken hoe hoog het suikergehalte is in je bloed. Dit gebeurt meestal doormiddel van een vingerprik. Als je nuchter bent hoort deze onder de 5.6 mmol/l te zijn. Na een maaltijd mag deze niet boven de 7.8 mmol/l komen. Als je nuchter bent en de waarde is 7 of hoger of als de waarde na een maaltijd 11 of hoger is, dan heb je diabetes. 

### Wat zijn de oorzaken? 

>Bij diabetes 2 is de oorzaak meestal overgewicht en het te veel hebben van buikvet. Ongezonde voeding, weinig lichaamsbeweging en roken zijn ook oorzaken zijn die een rol Spelen bij het ontstaan van diabetes 2. Er is ook een erfelijke mogelijkheid. Er is weinig bekend over de oorzaken van diabetes 1, een persoon heeft meestal aanleg voor diabetes 1 zonder dat dit perse in de familie zit. 

>Sommige virussen of bepaalde voeding 	voor of 	na de geboorte kunnen een rol spelen in het krijgen van diabetes 1, de darmflora heeft ook misschien een rol in het ontstaan van diabetes 1. 

### Hoe is het te behandelen? 
>Gezond eten, genoeg bewegen, niet roken, val af als nodig, slaap genoeg  
Medicijnen (die de cholesterol en bloeddruk verlagen)  
En je bloedsuiker meten om te checken hoeveel insuline er nodig is om jezelf gezond te houden 
(Dit gebeurt meestal via een insulinepomp) 

### Wat als je niet behandeld? 
> Als je diabetes niet behandeld, dan kan je last krijgen van meerdere symptomen, zoals veel dorst, veel plassen, wondjes die niet goed helen, een ziek gevoel en infecties. Bij het te weinig/veel innemen van suiker is er de mogelijkheid om in coma of zelfs dood te gaan. 

### Wat zijn de symptomen? 
>Als eerst kan je lichaam geen insuline meer aanmaken. Daardoor stapelt het suikergehalte in 	het bloed zich op, en kunnen die cellen niet meer die suiker gebruiken om energie op te 	wekken. Daardoor ontstaan er diverse symptomen:  

>Allereerst probeert je lichaam van dat suiker af te komen, dit wordt gedaan door middel van de plassen. Hierdoor probeert je lichaam de suiker te lossen door middel van de urine. Dit heeft als gevolg dat dit meestal gepaard gaat met het veel willen drinken om zo het vocht weer aan te vullen. Omdat je lichaam niet meer suiker kan gebruiken om energie te verwekken moeten ze deze energie uit andere bronnen halen. Hierdoor ontstaat er een versnelde vet verbranding waardoor degene ook snel af valt. Binnen 3 weken kan je tot wel 15 kilo afvallen. Bij vetverbranding komen helaas ook veel afvalstoffen vrij, en aangezien dit in een zo versneld proces gebeurt kan je lichaam deze afvalstoffen niet meer kwijtraken. Je adem gaat daardoor naar aceton ruiken. Dit is een erg belangrijk symptoom. Naast deze versnelde vetverbranding kan je lichaam ook minder aan energie komen, dit heeft als gevolg dat mensen met diabetes meestal snel moe/misselijk worden. 

> Verschil in symptomen diabetes type 1 ten opzichte van type 2: 

*   In type 2 zijn de wondjes slechter te genezen 
*   In type 2 zijn er vaak terugkerende infecties, zoals blaasontsteking.
*   In type 2 krijgen hebben vaker een droge mond 
*   In type 2 krijgen mensen vaker last van hun ogen, zoals rode ogen en wazig zien
